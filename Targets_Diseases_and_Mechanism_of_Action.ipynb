```python
moa = []
for i in tqdm_notebook(range (0, 2645)):
    exmpl_text = data.iloc[i, 1]
    doc = nlp(exmpl_text)
    drug_entities = [ent for ent in doc.ents if ent.label_ == "SIMPLE_CHEMICAL"]
    for ent in drug_entities:
        head = ent.root.head
        if head.lemma_ == ("inhibit|block"):
            targets = [token for token in head.children if token.ent_type_ == "GENE_OR_GENE_PRODUCT"]
            for target in targets:
                role = "inhibitor"
                MoA = f"{target} {role}"
                moa.append({
                    'Mechanism of Action': MoA,
                    'CID': data.iloc[i, 0]
                })
    target_entities = [ent for ent in doc.ents if ent.label_ == "GENE_OR_GENE_PRODUCT"]
    for ent in target_entities:
        head = ent.root.head
        if head.lemma_ == "inhibitor" or head.lemma_ == "Inhibitor" or head.lemma_ == "blocker" or head.lemma_ == "Blocker" or head.lemma_ == "antagonist" or head.lemma_ == "Antagonist" or head.lemma_ == "agonist" or head.lemma_ == "Agonist":
            role = head.lemma_
            MoA = f"{ent} {role}"
            moa.append({
                'Mechanism of Action': MoA,
                'CID': data.iloc[i, 0]
            })
```

```python
MoA = pd.DataFrame(moa)
MoA.head()
```
