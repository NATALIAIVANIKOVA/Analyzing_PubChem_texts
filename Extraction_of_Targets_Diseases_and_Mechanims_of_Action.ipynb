```python
import nltk
from nltk.stem import SnowballStemmer
stemmer = nltk.stem.SnowballStemmer('english')
```

```python
print (stemmer.stem ('antagonistic'))
```

```python
moa = []
for i in tqdm_notebook(range(bioactivity.shape[0])):
  exmpl_text = bioactivity.iloc[i, 2]
  doc = nlp(exmpl_text)
  drug_entities = [ent for ent in doc.ents if ent.label_ == "SIMPLE_CHEMICAL"]
  for ent in drug_entities:
      head = ent.root.head
      if head.lemma_ == ('inhibit|block'):
          targets = [token for token in head.children if token.ent_type_ == "GENE_OR_GENE_PRODUCT"]
          for target in targets:
              role = "inhibitor"
              MoA = f"{target} {role}"
              moa.append({
                  'Mechanism of Action': MoA,
                  'Activity': bioactivity.iloc[i, 1],
                  'CID': bioactivity.iloc[i, 0]
              })
  target_entities = [ent for ent in doc.ents if ent.label_ == "GENE_OR_GENE_PRODUCT"]
  for ent in target_entities:
      head = ent.root.head
      if head.lemma_ == "inhibitor" or head.lemma_ == "Inhibitor" or head.lemma_ == "blocker" or head.lemma_ == "Blocker" or head.lemma_ == "antagonist" or head.lemma_ == "Antagonist" or head.lemma_ == "agonist" or head.lemma_ == "Agonist" or head.lemma_ == "modulator" or head.lemma_ == "Modulator":
          role = head.lemma_
          MoA = f'{ent} {role}'
          moa.append({
            'Mechanism of Action': MoA,
            'Activity': bioactivity.iloc[i, 1],
            'CID': bioactivity.iloc[i, 0]
          })
      if head.lemma_ == "inhibition" or head.lemma_ == "Inhibition":
        role = 'inhibitor'
        MoA = f'{ent} {role}'
        moa.append({
          'Mechanism of Action': MoA,
          'Activity': bioactivity.iloc[i, 1],
          'CID': bioactivity.iloc[i, 0]
        })
      if head.lemma_ == "antagonism" or head.lemma_ == "Antagonism":
        role = 'antagonist'
        MoA = f'{ent} {role}'
        moa.append({
          'Mechanism of Action': MoA,
          'Activity': bioactivity.iloc[i, 1],
          'CID': bioactivity.iloc[i, 0]
        })
      if head.lemma_ == "activity":
        role_candidates = [token for token in head.children]
        for role_candidate in role_candidates:
          if role_candidate.lemma_ == 'antagonistic' or role_candidate.lemma_ == 'agonistic' or role_candidate.lemma_ == 'Antagonistic' or role_candidate.lemma_ == 'Agonistic':
            role = stemmer.stem(role_candidate.lemma_)
            MoA = f'{ent} {role}'
            moa.append({
              'Mechanism of Action': MoA,
              'Activity': bioactivity.iloc[i, 1],
              'CID': bioactivity.iloc[i, 0]
            })
          elif role_candidate.lemma_ == 'inhibitory':
            role = 'inhibitor'
            MoA = f'{ent} {role}'
            moa.append({
              'Mechanism of Action': MoA,
              'Activity': bioactivity.iloc[i, 1],
              'CID': bioactivity.iloc[i, 0]
            })
      if head.lemma_ == 'target':
        role_candidate = ent.root.head.head
        if role_candidate.lemma_ == 'inhibitor':
          role = role_candidate.lemma_
          MoA = f'{ent} {role}'
          moa.append({
            'Mechanism of Action': MoA,
            'Activity': bioactivity.iloc[i, 1],
            'CID': bioactivity.iloc[i, 0]
          })
```
